Q1. Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function. Upload a screenshot of the results and code from the beginning of Question 1 to the results below.

```python
import yfinance as yf

ticker_symbol_tesla = "TSLA"

tesla = yf.Ticker(ticker_symbol_tesla)
tesla_stock_data = tesla.history(period='max')
tesla_stock_data.reset_index(inplace=True)
tesla_stock_data.head()
```

Q5. Plot tesla stock graph, use make_graph function to graph the tesla stock data. also provide the title to the graph

```python
import yfinance as yf
import matplotlib.pyplot as plt

# stock_symbol = "TSLA"

# tesla_data = yf.download(stock_symbol, start="2023-01-01", end="2023-08-01")

def make_graph(data, title):
    plt.figure(figsize=(20, 6))
    plt.plot(data.index, data['Close'], label='Close Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

graph_title = "Tesla Stock Price"

make_graph(tesla_stock_data, graph_title)
plt.show()
```
